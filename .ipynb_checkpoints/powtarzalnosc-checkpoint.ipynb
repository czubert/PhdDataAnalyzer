### Powtarzalność, czyli sprawdzenie jak uśrednione widmo dla każdego podłoża różni się między podłożami

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/custom/saved_spectra_2.csv')
df.head(5)
```

```python
df = df.set_index('Raman Shift').T.iloc[:,-10:]
```

```python
df
```

```python
num_rows = len(df)
podloze = np.array([0] * (num_rows // 3) + [1] * (num_rows - 2*(num_rows // 3)) + [2] * (num_rows - 2*(num_rows // 3)))
df['podloze'] = podloze

# modyfikacja = np.tile([0, 1], int(len(df) / 2) + 1)[:len(df)]
# df['modyfikacja'] = modyfikacja
```

```python
df
```

```python
std = df.iloc[:,:].groupby(['podloze']).mean()
```

```python
std
```

```python
std.std()
```

```python
std.std().sum()
```
