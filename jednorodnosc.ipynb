### Jednorodność, czyli sprawdzenie jak bardzo kolejne pomiary na jednym podłożu różnią się od siebie

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/custom/saved_spectra_2.csv')
df.head(5)
```

```python
df = df.set_index('Raman Shift').T.iloc[:,-10:]
```

```python
df.head(5)
```

```python
num_rows = len(df)
podloze = np.array([0] * (num_rows // 2) + [1] * (num_rows - num_rows // 2))
modyfikacja = np.tile([0, 1], int(len(df) / 2) + 1)[:len(df)]
df['podloze'] = podloze
df['modyfikacja'] = modyfikacja
```

```python
df.head()
```

Liczymy odchylenie standardowe dla wszystkich pomiarów na danym podłożu. W wyniku tej operacji otrzymujemy odchylenia standardowe dla każdego punktu w obrębie pojedynczego podłoża.

```python
std = df.iloc[:,:].groupby(['modyfikacja', 'podloze']).std()
```

```python
std
```

Następnie liczymy średnią odchyleń dla wszystkich punktów w ramach jednego podłoża.

```python
n_features = std.shape[1]
std['std_mean'] = std.iloc[:,:n_features].mean(axis=1)
```

```python
std
```

Mamy średnie odchylenie dla każdego podłoża bez modyfikacji i z modyfikacją. Możemy teraz policzyć odchylenie/średnią z odchylenia dla wszystkich podłoży z modyfikacją i bez niej. Będzie to ostateczny wynik, który możemy użyć do porówniania między sobą podłoży.
